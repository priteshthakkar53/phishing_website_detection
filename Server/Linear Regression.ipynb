```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
phish = pd.read_csv('phishing_train.csv')
```

```python
phish.columns
```

For Phishing websites

```python
X_lr = phish[['Protocol', 'Length', '-', '@', 'Dots', 'In Domain Http',
       'Is IP Address', 'Harmful Host', 'Alexa Availability', 'Alexa Rank',
       'Whois Availability', 'Rank2traffic Availability',
       'Siterankdata Availability', 'Daily Unique Visitors']]
y_lr = phish['Result']
```

```python
phish.isnull().any()
```

```python
from sklearn.model_selection import train_test_split
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X_lr, y_lr, test_size=0.15)

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
classifier_lr = LinearRegression()
classifier_lr.fit(X_train_lr,y_train_lr)
predictions_lr = classifier_lr.predict(X_test_lr)
```

```python
accuracy_score(y_test_lr, predictions_lr)
```

```python
accuracy_lr = 100.0 * accuracy_score(y_test_lr, predictions_lr)
print("Linear Regression accuracy : " + str(accuracy_lr))
```
